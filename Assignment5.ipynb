1. Cleaning the Text:
Removing Punctuation: Punctuation marks are often irrelevant for many LSTM tasks and can sometimes interfere with the model's learning. so I remove the citation , Copyright Notice or Project Gutenberg License

```python
import requests
import re
## I am intersted in the "Persuasion" by Jane Austen Language: English
# Download the text
url = "https://www.gutenberg.org/cache/epub/105/pg105.txt"# Persuasion
response = requests.get(url)
text = response.text

# Remove headers and footers
start = text.find("*** START OF THE PROJECT GUTENBERG EBOOK PERSUASION ***") + len("*** START OF THE PROJECT GUTENBERG EBOOK PERSUASION ***")
end = text.find("*** END OF THE PROJECT GUTENBERG EBOOK PERSUASION ***")
text = text[start:end]

# Remove everything before the first chapter, which starts with "Chapter 1"
chapter_one_start = re.search(r"CHAPTER I.", text)
if chapter_one_start:
    text = text[chapter_one_start.start():]  # Extract from "CHAPTER I." onwards

# Remove punctuation and normalize whitespaces
text = re.sub(r'[^\w\s]', ' ', text)
text = re.sub(r'\s+', ' ', text).lower()

# Remove headers with titles and author names
text = re.sub(r'chapter i chapter ii chapter iii chapter iv chapter v chapter vi chapter vii chapter viii chapter ix chapter x chapter xi chapter xii chapter xiii chapter xiv chapter xv chapter xvi chapter xvii chapter xviii chapter xix chapter xx chapter xxi chapter xxii chapter xxiii chapter xxiv chapter i ', '', text)
# Remove common introductory sections such as "Preface," "Introduction," or "Foreword"
# use regex to find these sections and remove them. The `.*?` ensures non-greedy matching to the nearest chapter or title.
text = re.sub(r'(Preface|Introduction|Foreword).*?(CHAPTER I.\.|Chapter 1\.|First Chapter|chapter i.|Finis|finis)', r'\2', text, flags=re.DOTALL|re.IGNORECASE)

text = re.sub(r'finis', '', text)
print(text)  # Print all the TEXT to check the cleaning
```

```python
import tensorflow as tf
import re
import string

# Define the function to pad punctuation
def pad_punctuation(s):
    s = re.sub(f"([{string.punctuation}])", r" \1 ", s)
    s = re.sub(" +", " ", s)
    return s

# Apply the padding function to the entire combined text
padded_text = pad_punctuation(combined_text)

print(padded_text)
```

```python
# Extract a sample of 50000 characters from the processed text
sample_size = 50000
sample_text = padded_text[:sample_size]

print("Sample text extracted successfully.")
print(sample_text)
```

```python
# Convert to a Tensorflow Dataset
import tensorflow as tf

# Convert the text into a dataset of characters
chars = tf.data.Dataset.from_tensor_slices(list(sample_text))

# Batch the characters into sequences of 100 for example purposes
seq_length = 100  # Length of each sequence
sequences = chars.batch(seq_length+1, drop_remainder=True)

def split_input_target(chunk):
    input_text = chunk[:-1]  # All but the last character
    target_text = chunk[1:]  # All but the first character
    return input_text, target_text

# Prepare the dataset with input and target text
dataset = sequences.map(split_input_target)

# Batch and shuffle the dataset
batch_size = 64
buffer_size = 10000

dataset = dataset.shuffle(buffer_size).batch(batch_size, drop_remainder=True)

print("TensorFlow dataset created successfully.")
```

```python
import tensorflow as tf

# Create the TextVectorization layer
vectorize_layer = tf.keras.layers.TextVectorization(
    max_tokens=10000,
    output_mode='int',
    output_sequence_length=100)

# Dataset should be a batch of raw strings
text_dataset = tf.data.Dataset.from_tensor_slices([sample_text])
vectorize_layer.adapt(text_dataset)

# Now you can convert text to batches of vectorized data
def vectorize_text(input_text, target_text):  # Accept both input and target
    input_text = tf.expand_dims(input_text, -1)
    target_text = tf.expand_dims(target_text, -1)
    return vectorize_layer(input_text), vectorize_layer(target_text) # vectorize both

# Assuming 'dataset' contains raw texts batched appropriately
vectorized_dataset = dataset.map(vectorize_text)
# Show some mappings
vocab = vectorize_layer.get_vocabulary()
print("Vocabulary sample:", vocab[:20])
```

3. Create the Training Set

```python
import tensorflow as tf

# Define the function to prepare input and target pairs
def prepare_inputs(text):
    # Ensure the input text is in the correct shape
    text = tf.expand_dims(text, -1)
    # Tokenize the sentences
    tokenized_sentences = vectorize_layer(text)
    # Create input and target texts
    x = tokenized_sentences[:, :-1]  # All but the last token for input
    y = tokenized_sentences[:, 1:]   # All but the first token for target
    return x, y
```

```python
train_ds = text_dataset.map(prepare_inputs)
```

4. Build the LSTM

```python
import tensorflow as tf
import tensorflow.keras.layers as layers
import tensorflow.keras.models as models

def create_lstm_model(vocab_size, embedding_dim, lstm_units):
    # Define the input layer
    inputs = layers.Input(shape=(None,), dtype="int32")  # `None` allows variable sequence length

    # Add the embedding layer
    x = layers.Embedding(vocab_size, embedding_dim)(inputs)

    # Add the LSTM layer
    x = layers.LSTM(lstm_units, return_sequences=True)(x)

    # Add the dense output layer
    outputs = layers.Dense(vocab_size, activation="softmax")(x)

    # Create the model using the functional API
    lstm = models.Model(inputs, outputs)

    # Compile the model
    lstm.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return lstm

# Constants for the model
vocab_size = 10000  # Adjust based on your dataset's vocabulary
embedding_dim = 100
lstm_units = 64

# Create the LSTM model
lstm= create_lstm_model(vocab_size, embedding_dim, lstm_units)

# Print the model summary to verify its structure
lstm.summary()
```

5. Train the LSTM

```python
import tensorflow as tf
from tensorflow.keras import layers, models, losses, callbacks
import numpy as np

class TextGenerator(callbacks.Callback):
    def __init__(self, index_to_word, top_k=10):
        super().__init__()
        self.index_to_word = index_to_word
        self.word_to_index = {word: index for index, word in enumerate(index_to_word)}

    def sample_from(self, probs, temperature=1.0):
        probs = probs ** (1 / temperature)
        probs = probs / np.sum(probs)
        return np.random.choice(len(probs), p=probs), probs

    def generate(self, start_prompt, max_tokens=50, temperature=1.0):
        start_tokens = [self.word_to_index.get(x, 1) for x in start_prompt.split()]
        sample_token = None
        info = []
        while len(start_tokens) < max_tokens and sample_token != 0:
            x = np.array([start_tokens])
            y = self.model.predict(x, verbose=0)
            sample_token, probs = self.sample_from(y[0][-1], temperature)
            info.append({"prompt": start_prompt, "word_probs": probs})
            start_tokens.append(sample_token)
            # Check if the sample_token is within the bounds of index_to_word
            if 0 <= sample_token < len(self.index_to_word):
                start_prompt += " " + self.index_to_word[sample_token]
            else:
                # Handle out-of-bounds sample_token (e.g., replace with UNK)
                start_prompt += " [UNK]"  # Or break the loop
                #print(f"Warning: sample_token {sample_token} out of bounds")
                #break # Consider breaking if out-of-bounds is an error
        print(f"\ngenerated text:\n{start_prompt}\n")
        return info

    def on_epoch_end(self, epoch, logs=None):
        self.generate("sir walter elliot of kellynch hall in somersetshire was a man", max_tokens=50, temperature=1.0)
```

```python
index_to_word = {index: word for index, word in enumerate(vocab)}
text_generator = TextGenerator(index_to_word)
EPOCHS=50
# Fit the model
lstm.fit(
    train_ds,
    epochs=EPOCHS,
    callbacks=[text_generator],
)
```

```python
def print_probs(info, vocab, top_k=5):
    for i in info:
        print(f"\nPROMPT: {i['prompt']}")
        word_probs = i["word_probs"]
        p_sorted = np.sort(word_probs)[::-1][:top_k]
        i_sorted = np.argsort(word_probs)[::-1][:top_k]
        for p, i in zip(p_sorted, i_sorted):
            print(f"{vocab[i]}:   \t{np.round(100*p,2)}%")
        print("--------\n")
```
